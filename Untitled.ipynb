```python
import sys
from os import path
import numpy as np
from PIL import Image
import wikipedia
from wordcloud import WordCloud, STOPWORDS

# get path to script's directory
#currdir = path.dirname(__file__)

def get_wiki(query):
    # get best matching title for given query
    title = wikipedia.search(query)[0]

    # get wikipedia page for selected title
    page = wikipedia.page(title)
    return page.content


def create_wordcloud(text):
    # create numpy araay for wordcloud mask image
    mask = np.array(Image.open(path.join("/Users/ericchiyembekeza/Desktop/The Chuma Group/Qualic_Logo-removebg-preview.png")))

    # create set of stopwords
    stopwords = set(STOPWORDS)

    # create wordcloud object
    wc = WordCloud(background_color="white",
                   colormap='bwr',
                   max_words=400,
                   contour_color="black",
                   contour_width=3,
                   mask=mask,
                   stopwords=stopwords)

    # generate wordcloud
    wc.generate(text)

    # save wordcloud
    wc.to_file(path.join("qhwc.png"))


if __name__ == "__main__":
	# get query
	query = sys.argv[1]

	# get text for given query
	text = get_wiki(query)
	
	# generate wordcloud
	create_wordcloud(text)
    
create_wordcloud(get_wiki("Healthcare analytics"))
```
